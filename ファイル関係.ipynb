```python
# CSVにあるデータを追加してExcelに出力する
import csv

# 変数にファイルパスを格納することで可読性や保守性を向上させる(リファクタリング)
file_Path  = "/content/drive/MyDrive/Colab Notebooks/素材/Sample.csv"

# 追記する(aは追記モード)
with open(file_Path, 'a') as f:
    writer = csv.writer(f)

    # indexも配列として格納
    writer.writerow(['7','島根','そば'])
    writer.writerow(['8','広島','もんじゃ焼き'])

```

```python
# CSVデータをExcelに出力する(unnamed: 0は非表示)
import csv
import pandas as pd

# 変数にファイルパスを格納することで可読性や保守性を向上させる(リファクタリング)
file_Path  = "/content/drive/MyDrive/Colab Notebooks/素材/"

read_CsvPath = file_Path + "Sample.csv"     # 読み込むCSV
excel_FilePath = file_Path + "Sample.xlsx"  # 書き込むExcel

# CSV読み込み(unnamed: 0を非表示にする)
data = pd.read_csv(read_CsvPath,index_col = 0)

# Excel形式で出力(Sheet nameで出力先のシート名を決める)
data.to_excel(excel_FilePath, sheet_name='食べ物',index = False)

```
