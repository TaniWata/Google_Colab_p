```python
# 2つの配列を用いて条件ループをする
# やりたいこと : 2つの配列で同じ値があるか条件で見つける
# そして新しい配列(要素数は0とする)を作成し、その配列に格納していく

# 配列宣言
one_list = [1,3,5,6,7,9]
two_list = [0,1,6,3,9,2]
list_number = []    # 格納用の空の配列

#メイン処理
for i in range(len(one_list)):
  for j in range(len(two_list)):
    # 空の配列に追加していく
    if one_list[i] == two_list[j]: list_number.append(one_list[i])

# 出力
print(list_number)

```
