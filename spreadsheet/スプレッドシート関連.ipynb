```python
# URLからスプレッドシートを開く(driveをマウントしてからパスのコピーがおすすめ)
url = "https://docs.google.com/spreadsheets/d/1rY4qTY7-cjxLRXYEFpQtB-vAVYe2OR6cnm-LvCmVLpE/"
ss = gc.open_by_url(url)

# シート取得
# st = ss.get_worksheet(0)    # 1番目のシート
st = ss.worksheet('Sample') # シート名がSample

# 単一セルに書き込む(上書きされる)
# st.update_acell("A1", "test")
st.update_cell(8,1,"test")
st.update_cell(8,2,"test")

# 複数セルに書き込む(上書きされる)
cell_list = st.range('A9:B15')
for cell in cell_list:
  cell.value = 'test'
st.update_cells(cell_list)

# 最終行の取得
# list_of_lists = st.get_all_values()
# print(len(list_of_lists))

# 最終列の取得
# list_of_lists = st.get_all_values()
# print(len(list_of_lists[0]))    # 特定列
```
